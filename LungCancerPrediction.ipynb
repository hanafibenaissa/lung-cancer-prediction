```python
x = df.drop(columns='LUNG CANCER', axis=1)
y = df['LUNG CANCER']


```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
### Logistic Regression
```

```python
from sklearn import linear_model
LinReg=linear_model.LogisticRegression(C=0.01,solver='liblinear')
LinReg.fit(x_train,y_train)
```

```python
Predictions=LinReg.predict(x_test)
```

```python
Proba_Predictions=LinReg.predict_proba(x_test)
```

```python
from sklearn import metrics
LR_Accuracy_Score = metrics.accuracy_score(Predictions,y_test)
LR_JaccardIndex = metrics.jaccard_score(Predictions,y_test)
LR_F1_Score = metrics.f1_score(Predictions, y_test)
LR_Log_Loss = metrics.log_loss(y_test,Proba_Predictions)
Report = {'Metrics':['Accuracy score','Jaccard Index','F1 score','Log loss'],'Values':[LR_Accuracy_Score,LR_JaccardIndex,LR_F1_Score,LR_Log_Loss]}
df=pd.DataFrame(Report)
df
```

```python
### K-Nearest Neighbors
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
Acc=[]
neighbors=[i for i in range(1,31)]
for i in range(1,31):
    KNN = KNeighborsClassifier(n_neighbors=i).fit(x_train,y_train)
    KNN.fit(x_train,y_train)
    predictions=KNN.predict(x_test)
    LR_Accuracy_Score = metrics.accuracy_score(predictions,y_test)
    Acc.append(LR_Accuracy_Score)
```

```python
plt.plot(neighbors, Acc)
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy score')
plt.title('Accuracy Scores for Different Numbers of Neighbors (KNN)')
print("Best Value is: ",max(Acc))
print("Best number of neighbors to select is",(Acc.index(max(Acc)))+1)

```
